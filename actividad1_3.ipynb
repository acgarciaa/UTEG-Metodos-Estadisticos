## Obtención y limpieza de datos

### Paso 1. Obtención de los conjuntos de datos

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los conjuntos de datos
registros = pd.read_csv('registries.csv')
clientes = pd.read_csv('clients.csv')

print("Primeras filas del DataFrame 'registros':")
display(registros.head())

print("\nPrimeras filas del DataFrame 'clientes':")
display(clientes.head())
```

### Paso 2. Exploración

```python
# Para entender la estructura del conjunto y los tipos de las variables
print("Información del DataFrame 'registros':")
registros.info()

print("\nInformación del DataFrame 'clientes':")
clientes.info()

# Estadística descriptiva base por columna para identificar posibles problemas
print("\nEstadísticas descriptivas del DataFrame 'registros':")
display(registros.describe(include='all'))

print("\nEstadísticas descriptivas del DataFrame 'clientes':")
display(clientes.describe(include='all'))

# Para identificar categorías inconsistentes en 'clientes'
print("\nConteo de valores para la columna 'category' en 'clientes':")
display(clientes['category'].value_counts())

# Para identificar columnas con valores faltantes
print("\nValores faltantes por columna en 'registros':")
display(registros.isnull().sum())

print("\nValores faltantes por columna en 'clientes':")
display(clientes.isnull().sum())
```

### Paso 3. Limpieza

#### Corregir inconsistencias en variables categóricas

```python
# Reemplazar el texto mal escrito 'Silvr' por 'Silver'
clientes['category'] = clientes['category'].replace('Silvr', 'Silver')

print("Conteo de valores de 'category' después de la corrección:")
display(clientes['category'].value_counts())
```

#### Imputar valores faltantes

```python
# Obtener la categoría más frecuente (moda) para 'clientes'
moda_category_clientes = clientes['category'].mode()[0]

# Reemplazar las categorías faltantes por la moda
clientes['category'] = clientes['category'].fillna(moda_category_clientes)

print("Valores faltantes en 'clientes' después de imputar 'category':")
display(clientes.isnull().sum())

# Calcular la media del número de vistas para 'registros'
media_views_registros = registros['n_views'].mean()

# Reemplazar los valores faltantes por la media
registros['n_views'] = registros['n_views'].fillna(media_views_registros)

print("\nValores faltantes en 'registros' después de imputar 'n_views':")
display(registros.isnull().sum())

# Alternativa: eliminar los registros que tengan valores faltantes
# En este caso se crea un nuevo conjunto de datos, conservando el original.
# registros_sin_na = registros.dropna()
# print("\nRegistros sin valores NA:")
# display(registros_sin_na.head())

# También es posible eliminar filas con NA únicamente en una variable específica:
# registros_sin_na_n_views = registros.dropna(subset=['n_views'])
# print("\nRegistros sin valores NA en 'n_views':")
# display(registros_sin_na_n_views.head())
```

#### Transformar variables

```python
# Convertir las fechas y horas al formato fecha-hora
registros['entry_time'] = pd.to_datetime(registros['entry_time'], format="%d/%m/%Y %H:%M")
registros['exit_time'] = pd.to_datetime(registros['exit_time'], format="%d/%m/%Y %H:%M")

# Calcular el tiempo de conexión en segundos y convertir a horas para facilitar la interpretación
registros['time'] = (registros['exit_time'] - registros['entry_time']).dt.total_seconds() / 3600

print("Primeras filas de 'registros' con las columnas de tiempo transformadas y 'time' calculado:")
display(registros.head())
```

#### Detectar inconsistencias

```python
# Identificar registros cuyo tiempo de conexión es negativo
inconsistencias_tiempo = registros[registros['time'] < 0]
print("Registros con tiempo de conexión negativo:")
display(inconsistencias_tiempo)

# Registrar inconsistencia para revisión posterior con sistemas
registros['inconsistente'] = np.where(registros['time'] < 0, 'Si', 'No')

print("\nPrimeras filas de 'registros' con la columna 'inconsistente':")
display(registros.head())
```

#### Detectar y corregir valores extremos

```python
# Calcular el límite superior utilizando el criterio del IQR para 'n_views'
Q1_n_views = registros['n_views'].quantile(0.25)
Q3_n_views = registros['n_views'].quantile(0.75)
IQR_n_views = Q3_n_views - Q1_n_views

limSup_n_views = Q3_n_views + 1.5 * IQR_n_views

print(f"Q1 de n_views: {Q1_n_views}")
print(f"Q3 de n_views: {Q3_n_views}")
print(f"IQR de n_views: {IQR_n_views}")
print(f"Límite superior para n_views (criterio IQR): {limSup_n_views}")

print("\nRegistros donde 'n_views' excede el límite superior:")
display(registros[registros['n_views'] > limSup_n_views])

# Solo reemplazar los valores muy extremos por la media
# Se usa la media calculada previamente (media_views_registros)

registros['n_views'] = np.where(registros['n_views'] > 52,
                                media_views_registros,
                                registros['n_views'])

print("\nEstadísticas de 'n_views' después de la corrección de valores extremos (reemplazados por la media):")
display(registros['n_views'].describe())
```

## Procesamiento y análisis descriptivo de datos

### Paso 4. Procesamiento de datos

En esta sección se presentan algunas de las funciones más utilizadas de la librería `pandas` para manipular conjuntos de datos. El objetivo es construir un conjunto de datos listo para el análisis, a partir de los archivos de registros de navegación y de información de clientes.  

En particular, se aprenderá a:  

- Unir dos conjuntos de datos mediante una variable común.  
- Filtrar observaciones que cumplen una determinada condición.  
- Seleccionar únicamente las variables de interés para el análisis.    

Cada una de estas operaciones genera un nuevo conjunto de datos o una nueva vista de la información, sin modificar necesariamente los datos originales.

```python
# Separar una variable en dos columnas utilizando el guion como separador
clientes[['city', 'country']] = clientes['city_country'].str.split('-', expand=True)

print("Primeras filas de 'clientes' después de separar 'city_country':")
display(clientes.head())

# Volver a unir las columnas en una sola variable manteniendo también las variables separadas
# En pandas, la operación 'unite' es más un 'join' de strings, no es una 'unión' como en R tidyr
# Si se desea recrear la columna 'city_country' a partir de 'city' y 'country', se hace así:
clientes['city_country_reunited'] = clientes['city'] + '-' + clientes['country']

print("\nPrimeras filas de 'clientes' con la columna 'city_country_reunited':")
display(clientes.head())

# Unir la información de registros y clientes utilizando la variable ID.
# El resultado solo incluye los clientes que aparecen en ambos conjuntos de datos (inner join).
registros2 = pd.merge(registros, clientes, on='ID', how='inner')

print("\nPrimeras filas del DataFrame 'registros2' (unión de registros y clientes):")
display(registros2.head())

# Crear un nuevo conjunto de datos que contenga únicamente los clientes cuya categoría es 'Silver'.
registros2_silver = registros2[registros2['category'] == 'Silver']

print("\nPrimeras filas del DataFrame 'registros2_silver' (clientes con categoría 'Silver'):")
display(registros2_silver.head())

# Mostrar únicamente las variables ID, número de visitas y número de compras.
# .loc[] o seleccionar directamente las columnas no modifica el DataFrame original;
# simplemente presenta las columnas indicadas.
columnas_seleccionadas = registros2[['ID', 'n_views', 'n_buy']]

print("\nColumnas seleccionadas (ID, n_views, n_buy) de 'registros2':")
display(columnas_seleccionadas.head())
```

### Paso 5. Análisis descriptivo de datos

```python
print("Estadísticas descriptivas del DataFrame 'registros2':")
display(registros2.describe(include='all'))
```

Podemos resumir la información por categoría de cliente. La función `groupby()` agrupa las observaciones y `agg()` calcula estadísticas descriptivas para cada grupo.

```python
resumen_por_categoria = registros2.groupby('category').agg(
    total_login=('ID', 'count'), # 'n()' en R es 'count' en pandas para una columna específica
    total_compras=('n_buy', 'sum'),
    tiempo_promedio=('time', 'mean')
).reset_index()

print("Resumen por categoría de cliente:")
display(resumen_por_categoria)
```

Podemos visualizar las relaciones entre las variables mediante un gráfico de pares. La función `seaborn.pairplot()` permite explorar de forma simultánea la distribución de cada variable y su relación con las demás.

```python
sns.pairplot(registros2[['category', 'n_views', 'n_buy']], hue='category')
plt.suptitle('Gráfico de Pares de Category, n_views y n_buy', y=1.02) # Ajustar título
plt.show()
```

## Resumen Ejecutivo

Este análisis exploratorio de datos (EDA) se realizó sobre dos conjuntos de datos: `registros` (registros de navegación) y `clientes` (información de clientes). A continuación, se detallan los principales hallazgos y el proceso seguido:

### Limpieza y Transformación de Datos:
- **Inconsistencias Categóricas:** Se corrigió un error de tipografía en la categoría de cliente, reemplazando 'Silvr' por 'Silver' para estandarizar los datos.
- **Valores Faltantes:** Se imputaron los valores faltantes en la columna 'category' de `clientes` con la moda (la categoría más frecuente). Para 'n_views' en `registros`, los valores faltantes se reemplazaron con la media de la columna.
- **Transformación de Tiempo:** Las columnas 'entry_time' y 'exit_time' se convirtieron a formato datetime y se calculó la duración de la conexión ('time') en horas, facilitando el análisis temporal.
- **Detección de Inconsistencias:** Se identificaron y marcaron registros con tiempos de conexión negativos, lo que indica posibles errores en la captura de datos que requieren revisión.
- **Manejo de Outliers:** Se detectaron valores extremos en 'n_views' utilizando el método del rango intercuartílico (IQR) y se reemplazaron por la media para mitigar su impacto en los análisis posteriores.

### Procesamiento de Datos:
- **Separación y Unión de Columnas:** La columna 'city_country' en `clientes` se dividió en 'city' y 'country', y luego se demostró cómo se pueden volver a unir, mostrando flexibilidad en la manipulación de cadenas de texto.
- **Unión de Conjuntos de Datos:** Los DataFrames `registros` y `clientes` se unieron mediante un `inner join` utilizando la columna 'ID', creando un conjunto de datos consolidado (`registros2`) que combina la información de navegación con los detalles del cliente.
- **Filtrado:** Se creó un subconjunto de datos (`registros2_silver`) para clientes con categoría 'Silver', demostrando la capacidad de segmentación de datos.

- **Estadísticas Generales:** Un resumen estadístico de `registros2` proporcionó una visión general de las distribuciones de las variables numéricas y la composición de las categóricas.
